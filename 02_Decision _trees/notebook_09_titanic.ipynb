```python
max_depth_values = range(1,100)
```

```python
scores_data = pd.DataFrame()
```

```python
for max_depth in max_depth_values:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    temp_score_data = pd.DataFrame({'max_depth' : [max_depth],
                                   'train_score' : [train_score],
                                    'test_score' : [test_score]
                                   })
    scores_data = scores_data.append(temp_score_data)
```

```python
scores_data.head()
```

```python
scores_data_long = pd.melt(scores_data, id_vars=['max_depth'],
                          value_vars=['train_score', 'test_score'],
                          var_name='set_type',
                          value_name='score' 
                          )
```

```python
scores_data_long.head()
```

```python
scores_data_long.tail()
```

```python
sns.lineplot(x='max_depth', y='score', hue='set_type', data=scores_data_long);
```

## Кросс-валидация

```python
from sklearn.model_selection import cross_val_score
```

```python
scores_data = pd.DataFrame()

for max_depth in max_depth_values:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    
    mean_cross_val_score = cross_val_score(clf, X_train, y_train, cv=5).mean()
    
    temp_score_data = pd.DataFrame({'max_depth' : [max_depth],
                                   'train_score' : [train_score],
                                    'test_score' : [test_score],
                                    'cross_val_score' : [mean_cross_val_score]
                                   })
    scores_data = scores_data.append(temp_score_data)
```

```python
scores_data_long = pd.melt(scores_data, id_vars=['max_depth'],
                          value_vars=['train_score', 'test_score', 'cross_val_score'],
                          var_name='set_type',
                          value_name='score' 
                          )
```

```python
sns.lineplot(x='max_depth', y='score', hue='set_type', data=scores_data_long);
```

```python
scores_data_long[scores_data_long.set_type == 'cross_val_score'].head()
```

### Поиск оптимального набора параметров по сетке

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = tree.DecisionTreeClassifier()
clf
```

```python
parameters = {'criterion' : ['gini', 'entropy'], 
            'max_depth' : range(1,30)
             }
```

```python
grid_search_cv_clf = GridSearchCV(clf, parameters, cv=5)
```

```python
grid_search_cv_clf.best_params_
```

```python
best_clf = grid_search_cv_clf.best_estimator_
```

#### *Пороговые вероятности*

```python
y_pred_proba = best_clf.predict_proba(X_test)
```

```python
y_pred_proba[:5]
```

```python
pd.Series(y_pred_proba[:,1]).hist();
```

```python
y_pred = np.where(y_pred_proba[:,1] > 0.8 , 1 , 0)
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
y_pred = np.where(y_pred_proba[:,1] > 0.2 , 1 , 0)
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
np.unique(y_pred_proba[:,1])
```

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba[:,1])
roc_auc= auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange',
         lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
